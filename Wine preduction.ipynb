# Train test split

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn=KNeighborsRegressor(n_neighbors = 4)
```

```python
knn.fit(x_train ,y_train)
```

```python
y_pre=knn.predict(x_test).astype(int)
```
